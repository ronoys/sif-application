# Removing Columns and Separating Movies and Series
I decided to remove a lot of the non-numerical columns such as the links. I also divided the data into movies and series in order to identify trends for each one independently.

```python
df.rename(columns={'Series or Movie': 'SeriesOrMovie', 'Hidden Gem Score': 'HiddenGemScore', 'Country Availability': 'CountryAvailability',
                  'Netflix Release Date': 'NetflixReleaseDate', 'Production House': 'ProductionHouse', 'Netflix Link': 'NetflixLink',
                  'TMDb Trailer': 'TMDBTrailer', 'IMDb Link': 'IMDBLink', 'Trailer Site': 'TrailerSite'
                  ,'Release Date': 'ReleaseDate',},inplace=True)


# Removing trailer and other external links
df_adjusted = df.drop(['NetflixLink', "IMDBLink", "Image", "Poster", "TMDBTrailer", "TrailerSite"], axis=1)
df_adjusted = df_adjusted.drop(['Runtime', 'Director', 'Writer', 'Actors', 'Summary', 'NetflixReleaseDate'], axis=1)

# Movies only
movies = df_adjusted.loc[df_adjusted['SeriesOrMovie'] == 'Movie']

# Series only
series = df_adjusted.loc[df_adjusted['SeriesOrMovie'] == 'Series']
```

# Goal: Predicting Awards Received by a Movie

## Step One: Selecting Features

```python
# Removing Non-Numeric Features
movies = movies.drop(['ReleaseDate', 'ProductionHouse', 'Genre', 'Tags', 'Languages', 'SeriesOrMovie','CountryAvailability','View Rating','Title'], axis=1)
movies.info()
```

## Step Two: Filling in Missing Data Values
I decided to do this by substituting the mean of each feature in for missing values

```python
# Filling in Data Values

movies['HiddenGemScore'].fillna(movies['HiddenGemScore'].mean(), inplace=True)
movies['IMDb Score'].fillna(movies['IMDb Score'].mean(), inplace=True)
movies['Rotten Tomatoes Score'].fillna(movies['Rotten Tomatoes Score'].mean(), inplace=True)
movies['Metacritic Score'].fillna(movies['Metacritic Score'].mean(), inplace=True)
movies['Awards Received'].fillna(movies['Awards Received'].mean(), inplace=True)
movies['Awards Nominated For'].fillna(movies['Awards Nominated For'].mean(), inplace=True)
movies['Boxoffice'].fillna(movies['Boxoffice'].mean(), inplace=True)
movies['IMDb Votes'].fillna(movies['IMDb Votes'].mean(), inplace=True)
```

## Step Three: Scaling
I opted to use the MinMaxScaler from sklearn

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

movies=pd.DataFrame(scaler.fit_transform(movies), columns=movies.columns, index=movies.index) 
movies.head()
```

## Step Four: Splitting into Testing and Training Set
I used an 80/20 train test split using sklearn

```python
# Splitting 80/20 Train/Test
X = movies.drop('Awards Received',axis=1)
y = movies['Awards Received']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 50)
```

## Step Five: Ridge Regression and Initial Parameters List
I decided to use a linear ridge regression model based on the size of the dataset and the nature of the regression. I then found the default parameters from the model.

```python
# Getting Parameters from Original Model

from sklearn.linear_model import Ridge 
lr = Ridge()
lr.fit(X_train, y_train)

y_pred=lr.predict(X_test)

lr.get_params()
```

```python
# Calculating the r^2 value from the model
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

## Step Six: Hyperparameter Optimization
For this step, I used GridSearch with 10-fold cross val.

```python
from sklearn.model_selection import GridSearchCV

grid = {'alpha': [0, 1, 10, 100, 1000],
 'fit_intercept': [True, False],
 'max_iter': [1000, 5000, 10000, 15000],
 'solver': ['auto', 'svd', 'cholesky'],
 'tol': [0.001, .0001, .00001]
}

gridsearch = GridSearchCV(
    estimator = lr,
    param_grid = grid,
    cv = 10,
    verbose = 1
)

gridsearch.fit(X_train, y_train)
```

```python
# Outputting the results from GridSearch
param_df = pd.DataFrame(gridsearch.cv_results_)
param_df.head()
```

```python
# Calculating the best hyperparameters
gridsearch.best_params_
```

## Step Seven: Scoring
I found the max test score from grid search and also calculated the Mean Absolute Error.

```python
from sklearn.model_selection import cross_val_score


max_test_score = np.max(param_df["mean_test_score"])
print('Max Test Score: ',max_test_score)

scores = cross_val_score(lr, X_train, y_train, scoring='neg_mean_absolute_error', n_jobs=-1)
scores = np.absolute(scores)
print('Mean MAE: %.3f (%.3f)' % (np.mean(scores), np.std(scores)))
```

## Conclusion
The Mean MAE is quite small (0.017) which means that generally predicted values from the model will not vary too much from the "real" value. However, the max test score is only around 0.6145 which is quite low. This indicates that the model did not perform very well with the testing set. In the future, I would hope to improve my method of replacing null values and also try to optimize different regression models in order to improve this testing score.
